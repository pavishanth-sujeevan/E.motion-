```python
import pandas as pd

# Load the file (Kaggle path)
df = pd.read_csv('/kaggle/input/emotionsdata/emotions.csv')

# Show what we actually have
print("Columns in the dataset:", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())
print("\nShape:", df.shape)

# Use the real column names
TEXT_COLUMN = 'text'     # the raw text
LABEL_COLUMN = 'label'   # 0-5 numeric labels

print("\nUnique emotion labels (numeric):", sorted(df[LABEL_COLUMN].unique()))

# Optional: map numbers → emotion names for readability
emotion_map = {
    0: 'sadness',
    1: 'joy',
    2: 'love',
    3: 'anger',
    4: 'fear',
    5: 'surprise'
}
df['emotion_name'] = df[LABEL_COLUMN].map(emotion_map)

print("\nUnique emotions (named):", sorted(df['emotion_name'].unique()))

def simple_clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    text = text.translate(str.maketrans('', '', punctuation))
    text = ' '.join(text.split())
    return text.strip()

print("\nCleaning text...")
df['clean_text'] = df[TEXT_COLUMN].apply(simple_clean_text)

# Show before/after using correct columns
print("\nBefore vs After cleaning (first 8 rows):")
print(df[[TEXT_COLUMN, 'clean_text', LABEL_COLUMN, 'emotion_name']].head(8))

# Length stats
df['text_length'] = df['clean_text'].str.len()
print("\nText length statistics:")
print(df['text_length'].describe())

print("\n5 shortest cleaned texts:")
print(df.sort_values('text_length').head(5)[['clean_text', 'emotion_name']])

print("\n5 longest cleaned texts:")
print(df.sort_values('text_length', ascending=False).head(5)[['clean_text', 'emotion_name']])

# Save the result
df.to_csv('emotions_cleaned_pandas_only.csv', index=False)
print("\nSaved to: emotions_cleaned_pandas_only.csv")
```
