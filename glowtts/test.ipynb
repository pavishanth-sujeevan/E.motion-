```python
import IPython.display as ipd
from TTS.utils.synthesizer import Synthesizer
import numpy as np
import IPython.display as ipd
import torch

# 1. Replace with the EXACT path the console gave you
model_path = "C:\TTS-glowtts\Base_Model\model_file.pth"
config_path = "C:\TTS-glowtts\Base_Model\config.json"

vocoder_path="C:/Users/amawg/AppData/Local/tts/vocoder_models--en--ljspeech--hifigan_v2/model.pth"
voc_config_path= "C:/Users/amawg/AppData/Local/tts/vocoder_models--en--ljspeech--hifigan_v2/config.json"

# 2. Load the model
syn = Synthesizer(
    tts_checkpoint=model_path, 
    tts_config_path=config_path,
    vocoder_checkpoint=vocoder_path,   # Path to your vocoder
    vocoder_config=voc_config_path,
    use_cuda=True
)


syn.vocoder_model.cuda = True  # Ensure it's on GPU for speed
syn.vocoder_model.denoiser = True  # <--- THIS ENABLES THE DENOISER
# 3. Generate audio
wav = syn.tts(
    text="Hi, how are you. I think you are doing great.",
    noise_scale=0.35,
    length_scale=1.15,
    noise_scale_w=0.8
)


#syn.tts_model.inference_noise_scale = 0.1
#syn.tts_model.length_scale = 1.2


# 4. Play it right here in VS Code
#ipd.Audio(wav, rate=22050)
```

```python
import numpy as np
import scipy.io.wavfile as wavfile

# 1. Normalize the data (Crucial since your max is 825+)
# We convert to 16-bit PCM which is the standard for .wav files
#wav_norm = wav / np.max(np.abs(wav))
max_val = np.max(np.abs(wav))
if max_val > 0:
    # Scale so the peak is 0.95 instead of 1.0
    wav_norm = (wav / max_val)
else:
    wav_norm = wav
wav_int = (wav_norm * 32767).astype(np.int16)

# 2. Save to your current folder
file_name = "test_output.wav"
wavfile.write(file_name, 22050, wav_int)

print(f"✅ Success! File saved as: {file_name}")
print("Go to your folder and double-click 'test_output.wav' to play it.")
```
