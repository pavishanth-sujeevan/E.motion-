```python
import pandas as pd
import re

df = pd.read_csv('/kaggle/input/tweet-emotions/tweet_emotions.csv')

try:
    df = df[['tweet_id', 'sentiment', 'content']].copy()
except KeyError:
    print("Columns in file:", df.columns.tolist())
    print("Please replace 'tweet_id' with the actual ID column name (often 'id')")
    raise

df = df.dropna(subset=['content', 'sentiment'])
df = df.drop_duplicates(subset=['content'])
df['content'] = df['content'].str.strip()

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_content'] = df['content'].apply(clean_text)
df = df[df['cleaned_content'].str.len() > 0].copy()

final_df = df[['tweet_id', 'cleaned_content', 'sentiment']]

final_df.to_csv('cleaned_tweet_data.csv', index=False)

print("Rows:", len(final_df))
print(final_df.head(8))
print("\nSentiment counts:")
print(final_df['sentiment'].value_counts())
print("\nFile saved: cleaned_tweet_data.csv")
```
