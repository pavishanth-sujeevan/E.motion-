Resample 24000hz english audio to 22050hz audio

```python
import librosa
import soundfile as sf
import os

target_sr = 22050
source_folder = "C:/TTS-glowtts/XTTS_Dataset/wavs"
output_folder = "C:/TTS-glowtts/XTTS_Dataset/resample_wav"

for filename in os.listdir(source_folder):
    if filename.endswith(".wav"):
        # Load and resample
        y, sr = librosa.load(os.path.join(source_folder, filename), sr=target_sr)
        # Save back out
        sf.write(os.path.join(output_folder, filename), y, target_sr)
```

```python
import librosa
import os

dataset_path = r"C:/TTS-glowtts/XTTS_Dataset/wavs" 

for file in os.listdir(dataset_path):
    if file.endswith(".wav"):
        full_path = os.path.join(dataset_path, file)
        
        y, sr = librosa.load(full_path, sr=None)
        print(f"File: {file} | Sample Rate: {sr} | Channels: {y.ndim}")
        
        if sr != 22050: 
            print(f"⚠️ WARNING: {file} has a different sample rate!")
```
