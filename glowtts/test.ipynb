Testing for English Language

```python
import IPython.display as ipd
from TTS.utils.synthesizer import Synthesizer
import numpy as np
import IPython.display as ipd
import torch

model_path = r"C:\TTS-glowtts\output\glow-tts-rel_pos_transformer-February-22-2026_06+45PM-0000000\best_model.pth"
config_path = r"C:\TTS-glowtts\output\glow-tts-rel_pos_transformer-February-22-2026_06+45PM-0000000\config.json"

vocoder_path="C:/Users/amawg/AppData/Local/tts/vocoder_models--en--ljspeech--hifigan_v2/model.pth"
voc_config_path= "C:/Users/amawg/AppData/Local/tts/vocoder_models--en--ljspeech--hifigan_v2/config.json"

#  Load the model
syn = Synthesizer(
    tts_checkpoint=model_path, 
    tts_config_path=config_path,
    vocoder_checkpoint=vocoder_path,   
    vocoder_config=voc_config_path,
    use_cuda=True
)


syn.vocoder_model.cuda = True  
syn.vocoder_model.denoiser = True  

#  Generate audio
wav = syn.tts(
    text="if you feel that you'd like to tell me what has hurt you, i'll be glad to listen.",
    noise_scale=0.35,
    length_scale=1.15,
    noise_scale_w=0.8
)


```

```python
import numpy as np
import scipy.io.wavfile as wavfile

max_val = np.max(np.abs(wav))
if max_val > 0:
    
    wav_norm = (wav / max_val)
else:
    wav_norm = wav
wav_int = (wav_norm * 32767).astype(np.int16)

# Save to current folder
file_name = "test_output.wav"
wavfile.write(file_name, 22050, wav_int)

print(f" Success! File saved as: {file_name}")
print("Go to your folder and double-click 'test_output.wav' to play it.")
```

```python
import numpy as np

print(f"Audio length: {len(wav)} samples")
print(f"Max amplitude: {np.max(np.abs(wav))}")
```

MCD Score for english

```python
from pymcd.mcd import Calculate_MCD

mcd_toolbox = Calculate_MCD(MCD_mode="dtw")

#  Paths to audio
ref_audio = r"C:\TTS-glowtts\XTTS_Dataset\wavs\5558.wav"
synth_audio = r"C:\TTS-glowtts\test_output.wav"

#  Calculate the score
mcd_value = mcd_toolbox.calculate_mcd(ref_audio, synth_audio)

print(f"Mel-Cepstral Distortion (MCD): {mcd_value:.4f}")
```
