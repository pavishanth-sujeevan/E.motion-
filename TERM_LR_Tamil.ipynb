```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.pipeline import Pipeline
```

```python
# Change path according to your location
df = pd.read_csv("/Users/bimsaraserasinghe/Downloads/DSGP/Pre-Processed Csvs/balanced_emotion_dataset (3).csv")

print("Dataset shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nUnique emotions:", sorted(df['Emotion'].unique()))
print("\nClass distribution:\n", df['Emotion'].value_counts())
print("\nMissing values:\n", df.isnull().sum())

# We will use Clean_Text and Label (already numeric 0–10)
X = df['Clean_Text'].astype(str)
y = df['Label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    stratify=y,              # important even when balanced
    random_state=42
)

print(f"Training samples : {len(X_train):,}")
print(f"Test samples      : {len(X_test):,}")
print(f"Number of classes : {len(np.unique(y))}")
```

```python
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        max_features=18000,       # good range for Tamil social media
        ngram_range=(1, 2),       # unigrams + bigrams → strong improvement
        min_df=3,
        max_df=0.95,
        analyzer='word',
        token_pattern=r'(?u)\b\w\w+\b',
        strip_accents='unicode'
    )),
    ('clf', LogisticRegression(
        multi_class='multinomial',
        solver='saga',
        C=1.5,                    # slightly higher than default
        max_iter=2000,
        random_state=42,
        n_jobs=-1
    ))
])

print("Pipeline created successfully")
```
