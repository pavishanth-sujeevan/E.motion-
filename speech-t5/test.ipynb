```python
import os
import torch
import soundfile as sf
from transformers import SpeechT5ForTextToSpeech, SpeechT5HifiGan, SpeechT5Processor

# 1. SET PATHS 
MODEL_CHECKPOINT = r"C:\TTS-Speech_t5\speecht5_finetuned_results\checkpoint-1000"

PROCESSOR_PATH = r"C:\TTS-Speech_t5\speecht5_tts"

VOCODER_PATH = r"C:\TTS-Speech_t5\speecht5_hifigan" 

device = "cuda" if torch.cuda.is_available() else "cpu"

# 2. LOAD COMPONENTS
print(f"Loading from: {MODEL_CHECKPOINT}")

# Load Processor (Handles text to IDs)
processor = SpeechT5Processor.from_pretrained(PROCESSOR_PATH)

# Load Model (The fine-tuned brain)
my_model = SpeechT5ForTextToSpeech.from_pretrained(MODEL_CHECKPOINT).to(device)

# Load Vocoder (Turns spectrograms into sound)
vocoder = SpeechT5HifiGan.from_pretrained(VOCODER_PATH).to(device)

# 3. INFERENCE
test_text = "if you feel that you'd like to tell me what has hurt you i'll be glad to listen."
inputs = processor(text=test_text, return_tensors="pt").to(device)

# Using the Zero embedding 
spk_embeddings = torch.zeros((1, 512)).to(device)

my_model.eval()
with torch.no_grad():
    speech = my_model.generate_speech(inputs["input_ids"], spk_embeddings, vocoder=vocoder)

# 4. SAVE
output_path = os.path.join(os.getcwd(), "v_code_test.wav")
sf.write(output_path, speech.cpu().numpy(), samplerate=16000)
print(f"Success! Audio saved to: {output_path}")
```
